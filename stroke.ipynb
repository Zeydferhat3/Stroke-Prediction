```python
data=data.drop(["id","gender","ever_married","Residence_type"], axis=1)
```

```python
data=data.dropna(subset=["bmi"])  
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
 
le = LabelEncoder()
 
smoking_status1 = le.fit_transform(data['smoking_status'])
work_type1 = le.fit_transform(data['work_type'])

data["smoking_status"] = smoking_status1
data["work_type"] = work_type1
```

```python
data
```

```python
corr_matrix =data.corr()
corr_matrix
```

```python
X =  data.drop(['stroke'],axis=1)
y = data['stroke']
```

# 3.LogisticRegression 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="liblinear")
model.fit(X_train, y_train)
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# 4.support vector machine

```python
from sklearn import svm
from sklearn import metrics

clf = svm.SVC(kernel="linear")
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

acc = metrics.accuracy_score(y_test, y_pred)
print(acc)
```
